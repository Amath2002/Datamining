**<h1 style="color:#06344ef6;text-align:center;font-size:35px;font-family:Bodoni MT">Modélisation</h1>**

```python
#Séparer la variable cible 'Attrition_Flag'  et les variables explicatives
y = df['Attrition_Flag']
x = df.drop('Attrition_Flag', axis = 1)
```

```python
#Subdivision du jeu de données en apprentissage et en test
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size =0.2, random_state = 2)
```

```python
#Vérifier le résultat
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
test = SelectKBest(f_classif, k=12)
```

```python
kbest = test.fit(x_train, y_train)
```

```python
kbest.get_feature_names_out()
```

```python
x_train=kbest.transform(x_train)
```

```python
x_train[1,:]
```

```python
#Création d'un objet lr de la classe LogisticRegression
lr = LogisticRegression(solver ='newton-cg', random_state =1)
```

```python
#Apprentissage du modèle
model = lr.fit(x_train, y_train)
```

```python
x_test=kbest.transform(x_test)
```

```python
#Application du modèle au données de test
y_pred = model.predict(x_test)
```

```python
#Taux de bonnes prédictions
accuracy = accuracy_score(y_test, y_pred)
accuracy
#Conclusion : le modèle produit 89.98 % de bonne prédiction.
```

```python
#Score d'entrainement (le taux de bonne prédiction calculé sur la base des données d'entrainement)
model.score(x_train, y_train)
```

```python
#Score de test (le taux de bonne prédiction calculé sur la base des données de test)
model.score(x_test, y_test)
```
